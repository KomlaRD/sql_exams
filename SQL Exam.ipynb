### Question 1

### Question 5

How many movies are there that are both in the "Thriller" genre and contain the word “love” anywhere in the keywords?


**Options:**
 - 48
 - 38
 - 14
 - 1

```python
%%sql
SELECT
    COUNT(m.movie_id) AS "Movie count",
    g.genre_name AS "Name of Genre",
    k.keyword_name AS "Name of Keyword"
FROM
    movies m
JOIN
    genremap gm
ON
    m.movie_id = gm.movie_id
JOIN
    genres g
ON
    g.genre_id = gm.genre_id
JOIN
    keywordmap km
ON
    m.movie_id = km.movie_id
JOIN
    keywords k
ON
    k.keyword_id = km.keyword_id
WHERE
    g.genre_name = "Thriller"
AND
    k.keyword_name = "love"
GROUP BY
    g.genre_name AND k.keyword_name;
```

### Question 6

How many movies are there that were released between 1 August 2006 ('2006-08-01') and 1 October 2009 ('2009-10-01') that have a popularity score of more than 40 and a budget of less than 50 000 000?

### Question 9

What are the three production companies that have the highest movie popularity score on average, as recorded within the database?


**Options:**

 - MCL Films S.A., Turner Pictures, and George Stevens Productions
 - The Donners' Company, Bulletproof Cupid, and Kinberg Genre
 - Bulletproof Cupid, The Donners' Company, and MCL Films S.A
 - B.Sting Entertainment, Illumination Pictures, and Aztec Musique

```python
%%sql
SELECT
    p.production_company_name,
    m.popularity,
    AVG(m.popularity) OVER (PARTITION BY p.production_company_name) AS "Average populariy score"
FROM
    movies m
JOIN
    productioncompanymap pm
ON
    m.movie_id = pm.movie_id
JOIN
    productioncompanies p
ON
    p.production_company_id = pm.production_company_id
ORDER BY "Average populariy score" DESC
```

### Question 10

How many female actors (i.e. gender = 1) have a name that starts with the letter "N"?


**Options:**

 - 0
 - 355
 - 7335
 - 1949

```python
%%sql
SELECT
    gender AS "Gender of actor",
    COUNT(gender) AS "Number of female actors with name beginning with N"
FROM
    actors
WHERE 
    gender = 1 -- Female
AND
    actor_name LIKE "N%" -- Name begins with N
GROUP BY
    gender
LIMIT
    10;
```

### Question 11

Which genre has, on average, the lowest movie popularity score? 


**Options:**

 - Science Fiction
 - Animation
 - Documentary
 - Foreign

```python
%%sql
SELECT
    g.genre_name,
    gm.genre_id,
    m.popularity,
    AVG(m.popularity) OVER (PARTITION BY g.genre_name) AS "Average score"
FROM
    movies m
JOIN
    genremap gm
ON
    m.movie_id = gm.movie_id
JOIN
    genres g
ON
    g.genre_id = gm.genre_id
ORDER BY "Average score"
LIMIT
    10;
```

### Question 12

Which award category has the highest number of actor nominations (actors can be male or female)? (Hint: `Oscars.name` contains both actors' names and film names.)

**Options:**

- Special Achievement Award
- Actor in a Supporting Role
- Actress in a Supporting Role
- Best Picture



```python
%%sql
SELECT
    name AS "Name of actor",
    film AS "Movie Title",
    COUNT(award) AS "Number of nominations",
    award AS "Award category"
FROM
    oscars
GROUP BY
    award
ORDER BY "Number of nominations" DESC
LIMIT
    10;
```
